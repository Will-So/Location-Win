In this notebook, I model LDA topics.

```python
from __future__ import division

import pandas as pd
import pyLDAvis.gensim
import gensim

pyLDAvis.enable_notebook()
```

```python
dictionary = gensim.corpora.Dictionary.load('../data/words.dict')
corpus = gensim.corpora.MmCorpus('../data/deerwester.mm')
lda = gensim.models.ldamodel.LdaModel.load('../lda_models/all_lda')
```

```python
pyLDAvis.gensim.prepare(lda, corpus, dictionary)
```
